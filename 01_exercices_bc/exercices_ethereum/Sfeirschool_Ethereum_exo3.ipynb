<h4> <b>What's smart contract structure?</b></h4> A contract can contain :
<ul>
    <li><b style="color:red">pragma version</b> The version used by compiler (Mandatory)</li>
    <li><b style="color:red">contract</b> The contract defines by 'contract' keyworld. (Mandatory)</li> 
    <li><b style="color:red">state variables</b> Value stores in smart contract.</li> 
    <li><b style="color:red">functions</b> Behavior which can be called by transactions.</li> 
    <li><b style="color:red">events</b> Way to communicate with EVM.</li> 
    <li><b style="color:red">struct type</b> Complex data type.</li> 
    <li><b style="color:red">More...</b> Function Modifiers, Enum types etc...</li> 
</ul>

<h4> <b>State variables</b></h4>
<ul>
    <li><b style="color:red">syntaxe</b> <code>[type] [scope] [reference]</code></li>
    <li><b style="color:red">type</b> int8, uint8, address...</li>
    <li><b style="color:red">scope</b> private, public....</li> 
</ul>
<i style="color:yellow">Do it yourself !!!</i>

Declare two variables:
<lu>
    <li><b style="color:red">(1)</b> Create a client counter, and contract owner.</li>
    <li><b style="color:red">(2)</b> Declare variable: <pre><code>uint8 private clientCount; // define a counter for bank clients.
address public owner; // define the owner of contract. type is address
</code></pre></li>
    <li><b style="color:red">(3)</b> Adapt constructor: <pre><code>    constructor() public payable {
        // set default value.
        clientCount = 0;
        owner = msg.sender; // msg allows to have details from Tx.
    }
</code></pre></li>
</lu>

<ul>

    <li><b style="color:red">sent event</b> <code>emit [name]([arguments])</code></li>

<i style="color:yellow">Do it yourself !!!</i>

<h4> <b>Functions</b></h4>
<ul>
    <li><b style="color:red">syntaxe</b> <code>function [name]([arguments])[scope][returns]</code></li>
</ul>

<i style="color:yellow">Do it yourself !!!</i>

Declare four function:
<lu>
    <li><b style="color:red">(1)</b> Create <i style="color:yellow">deposit()</i> (Get account and deposit, to store on balance),<i style="color:yellow">withdraw()</i> (Get money from account), <i style="color:yellow">balance()</i> (Return current balance of account) functions.</li>
    <li><b style="color:red">(2)</b> Declare function: <pre><code>//deposit
    function deposit(address adr, uint value) public payable returns (uint) {
    balances[adr] += value;
    emit WhoAmI(adr, value);
    return balances[adr];
}
//withdraw
function withdraw(address from, address to, uint withdrawAmount) public returns (uint remainingBal) {
        emit WhoAmI(from, balances[from]);
        emit WhoAmI(to, balances[to]);
        if (withdrawAmount <= balances[from]) {
            balances[from] -= withdrawAmount;
            to.transfer(withdrawAmount);
        }
        return balances[from];
    }
// balance
function balance(address adr) public constant returns (uint) {
    return balances[adr];
}</code></pre></li>
</lu>
